### ¬øQu√© hace pd.read_csv?


#### Descripci√≥n t√©cnica:
pd.read_csv() es una funci√≥n de alto nivel provista por la biblioteca Pandas. Su prop√≥sito es leer archivos en formato CSV (Comma-Separated Values) y convertirlos en un objeto DataFrame, que es una estructura tabular con filas y columnas, similar a una hoja de c√°lculo o una tabla SQL.


#### üîß ¬øQu√© hace internamente?
* Abre el archivo CSV especificado en la ruta proporcionada (filepath_or_buffer).
* Lee el contenido l√≠nea por l√≠nea, separando los valores por delimitadores (por defecto, la coma ,).
* Asigna la primera fila como encabezado (header=0 por defecto), a menos que se indique lo contrario.
* Convierte cada columna en una Series de Pandas con un tipo de dato inferido (int, float, string, bool, etc.).
* Devuelve un objeto DataFrame, optimizado para an√°lisis y manipulaci√≥n en memoria.


#### Significado detallado:
* 'data/batallas.csv' es la ruta relativa al archivo CSV que contiene los datos de las batallas (por ejemplo: nombre, a√±o, ej√©rcitos, comandantes, resultados, etc.).
* pd.read_csv(...) lee el archivo ubicado en esa ruta y lo convierte en un DataFrame.
* batallas = ... asigna ese DataFrame a la variable batallas, que queda disponible en memoria para posteriores an√°lisis, visualizaci√≥n o modelado.

```python
batallas = pd.read_csv('data/batallas.csv')  # Carga un archivo CSV y lo convierte en un DataFrame de Pandas
```

### ¬øQu√© hace .value_counts() en Pandas?

#### Descripci√≥n t√©cnica:
.value_counts() es un m√©todo de objetos Series en Pandas que devuelve una Serie con la frecuencia de cada valor √∫nico, es decir, cu√°ntas veces aparece cada valor distinto.

    * Ordena el resultado de mayor a menor (por defecto).
    * Devuelve el resultado como una nueva Series, con el valor como √≠ndice y la frecuencia como valor.


* Argumentos √∫tiles:
    * normalize=True: en lugar del conteo absoluto, devuelve la proporci√≥n (frecuencia relativa).
    * sort=False: evita el ordenamiento autom√°tico.
    * ascending=True: ordena de menor a mayor.
    * dropna=False: incluye los valores NaN en el conteo.

### 1. .notnull()

#### Descripci√≥n t√©cnica:
M√©todo de Pandas que devuelve una Serie booleana con True en las posiciones donde el valor no es nulo (NaN).
#### üîß ¬øPara qu√© sirve?
Permite filtrar registros v√°lidos en una columna antes de realizar transformaciones que no aceptan NaN, como convertir tipos o aplicar funciones matem√°ticas.


### 2. .copy()

#### Descripci√≥n t√©cnica:
Crea una copia independiente del subconjunto del DataFrame. Es crucial para evitar advertencias (SettingWithCopyWarning) cuando se modifican datos despu√©s de un filtrado.
#### ¬øPor qu√© es importante?
Si no se usa .copy(), Pandas puede modificar la vista original de decesos, lo que puede generar resultados inesperados.


### 3. .astype(int)

#### Descripci√≥n t√©cnica:
Convierte expl√≠citamente los datos de la columna al tipo de dato entero (int). Si los datos vienen como float64 (ej: 299.0), esta conversi√≥n los transforma en enteros (299).
* üîß ¬øPor qu√© se usa?
    * Mejora la limpieza del dataset.
    * Reduce el consumo de memoria.
    * Evita errores en visualizaci√≥n (por ejemplo, ejes con decimales innecesarios en seaborn).

Requiere que no haya NaN, por eso se hace despu√©s de .notnull().


### 4. .unique()

Descripci√≥n t√©cnica:
Devuelve un ndarray de NumPy con todos los valores √∫nicos de una columna (sin duplicados), en el orden en que aparecen.

#### üîß ¬øPara qu√© sirve?
* Ideal para:
    * Verificar cu√°ntos a√±os distintos hay.
    * Saber si los datos ya est√°n limpios.
    * Usar los valores √∫nicos como etiquetas para visualizaci√≥n, agrupamiento, etc.


### 5. plt.title()

#### Descripci√≥n t√©cnica:
Asigna un t√≠tulo principal al gr√°fico actual generado por Matplotlib.
* ‚öôÔ∏è Argumentos principales:
    * label ‚Üí texto del t√≠tulo (string).
    * fontsize, fontweight, loc ‚Üí personalizaci√≥n del estilo del t√≠tulo.


### 7. plt.xlabel()

#### Descripci√≥n t√©cnica:
Etiqueta el eje horizontal (X) del gr√°fico con texto personalizado.


### 8. plt.ylabel()

#### Descripci√≥n t√©cnica:
Etiqueta el eje vertical (Y) del gr√°fico con una descripci√≥n textual.

### Objetivo del an√°lisis:

Contar cu√°ntos personajes aparecen en cada uno de los libros de la saga:

* Game of Thrones: `game_of_thrones`
* Clash of Kings: `clash_of_kings`
* Storm of Swords: `storm_of_words`
* Feast for Crows: `feast_for_crows`

### Objetivo pr√≥xima celda de c√≥digo:

Calcular **cu√°ntos cap√≠tulos pasan entre la primera aparici√≥n del personaje y su muerte** dentro del mismo libro. Esto nos da una idea del **"tiempo narrativo de vida"** de cada personaje que muri√≥.

```python
# Filtramos personajes que tengan tanto cap√≠tulo de introducci√≥n como cap√≠tulo de muerte
fallecidos = decesos[
    decesos['capitulo_muerte'].notnull() & 
    decesos['capitulo_intro'].notnull()
].copy()

# Calculamos la "duraci√≥n narrativa" en cap√≠tulos
fallecidos['capitulos_antes_de_morir'] = (
    fallecidos['capitulo_muerte'] - fallecidos['capitulo_intro']
).astype(int)

# Estilo general
sns.set_theme(style='whitegrid', palette='pastel')

# Visualizaci√≥n 1: Histograma
plt.figure(figsize=(8, 5))
sns.histplot(fallecidos['capitulos_antes_de_morir'], bins=20, kde=True)
plt.title('Distribuci√≥n de cap√≠tulos entre introducci√≥n y muerte')
plt.xlabel('Cap√≠tulos antes de morir')
plt.ylabel('Cantidad de personajes')
plt.tight_layout()
plt.show()

# Visualizaci√≥n 2: Boxplot para ver mediana y outliers
plt.figure(figsize=(10, 7))
sns.boxplot(x=fallecidos['capitulos_antes_de_morir'], color='skyblue')
plt.title('Boxplot: cap√≠tulos antes de morir')
plt.xlabel('Cap√≠tulos antes de morir')
plt.tight_layout()
plt.show()
```

### Objetivo pr√≥xima celda:

Identificar qu√© **casas** (facci√≥n o lealtad de los personajes) tienen **m√°s personajes activos en los libros 4 y 5**, que son los m√°s avanzados cronol√≥gicamente.

Esto nos da una idea de:

* Qu√© casas mantienen relevancia en la narrativa a lo largo del tiempo.
* Posible protagonismo o supervivencia de ciertas casas.

```python
# Eliminamos personajes sin casa definida para evitar 'NaN' o nulos
decesos_filtrados = decesos[decesos['lealtad_a_casa'].notnull()].copy()

# Agrupamos por casa y sumamos cu√°ntos personajes aparecen en libros 4 (FfC) y 5 (DwD)
apariciones_posteriores = decesos_filtrados.groupby('lealtad_a_casa')[['feast_for_crows', 'dance_with_dragons']].sum()

# Quitamos casas con muy pocas apariciones totales para evitar ruido visual
apariciones_filtradas = apariciones_posteriores[apariciones_posteriores.sum(axis=1) > 1]

# Configuramos el estilo visual
sns.set_theme(style='white', palette='pastel')

# Visualizamos como heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(apariciones_filtradas, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)

# Personalizamos con algunos t√≠tulos
plt.title('Apariciones por casa en libros 4 y 5 (FfC, DwD)')
plt.xlabel('Libro')
plt.ylabel('Casa')
plt.tight_layout()

# Mostramos el gr√°fico
plt.show()
```

### Objetivo pr√≥xima celda:

Analizar si existe una diferencia significativa entre **g√©neros** en cuanto al **n√∫mero de libros en los que aparece un personaje**.

Esto puede revelar:

* ¬øQu√© g√©nero tiene mayor "presencia narrativa"?

```python

# Sumamos la aparici√≥n en libros por personaje (games_of>_thrones a dance_with_dragons = columnas binarias con 1 o 0)
libros = ['game_of_thrones','clash_of_kings','storm_of_words','feast_for_crows','dance_with_dragons']
decesos['total_libros'] = decesos[libros].sum(axis=1)

# Filtramos filas con valores v√°lidos en la columna 'genero'
# En el dataset 0 = mujer, 1 = hombre
decesos_validos = decesos[decesos['genero'].isin([0, 1])].copy()

# Estilo del gr√°fico
sns.set_theme(style='whitegrid', palette='pastel')

# Gr√°fico violinplot
plt.figure(figsize=(9,5))
sns.violinplot(x='genero', y='total_libros', data=decesos_validos)

# Algunos t√≠tulos descrptivos
plt.title('Distribuci√≥n de libros en los que aparece un personaje seg√∫n g√©nero')
plt.xlabel('G√©nero (0 = mujer, 1 = hombre)')
plt.ylabel('Cantidad de libros en los que aparece')
plt.tight_layout()

# Y mostramos el gr√°fico
plt.show()
```

## ü§ñ Una cela de c√≥digo predicci√≥n con regresi√≥n log√≠stica: ¬øPodemos predecir si un personaje muere?

### üéØ Objetivo:
Construir un **modelo de clasificaci√≥n binaria** para predecir si un personaje muere (`1`) o sobrevive (`0`), usando variables como g√©nero, nobleza y aparici√≥n en libros.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# PASO 1: preparaci√≥n de los datos
# Creamos la variable objetivo: 1 si muri√≥, 0 si no muri√≥
decesos['murio'] = decesos['anio_muerte'].notnull().astype(int)

# Seleccionamos caracter√≠sticas relevantes
caracteristicas = ['genero', 'nobleza', 'game_of_thrones','clash_of_kings','storm_of_words','feast_for_crows','dance_with_dragons', 'capitulo_intro']
decesos = decesos[caracteristicas + ['murio']].dropna()  # Quitamos filas con NaN

# X: variables independientes (vector de caracter√≠sticas), y: variable objetivo ( y predicha)
X = decesos[caracteristicas]
y = decesos['murio']

# Paso 2: divisi√≥n de los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# PASO 3: escalado (opcional, m√°s √∫til para regresi√≥n log√≠stica)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Paso 5: aplicamos el modelo de Regresi√≥n Log√≠stica
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)

print('Reegeresi√≥n log√≠stica')
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))

# Paso 6: aplicamos el modelo: Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print('Random Forest')
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

## Interpretaci√≥n de salida
* **Precision**: Qu√© tan bien predice las muertes verdaderas.
* **Recall**: Qu√© tanto de las muertes reales fueron detectadas.
* **F1-score**: Media arm√≥nica de precision y recall.
* **Confusion matrix**: muestra verdaderos positivos, falsos negativos, etc.
