### 🧠 ¿Qué hace pd.read_csv?


#### 📌 Descripción técnica:
pd.read_csv() es una función de alto nivel provista por la biblioteca Pandas. Su propósito es leer archivos en formato CSV (Comma-Separated Values) y convertirlos en un objeto DataFrame, que es una estructura tabular con filas y columnas, similar a una hoja de cálculo o una tabla SQL.


#### 🔧 ¿Qué hace internamente?
* Abre el archivo CSV especificado en la ruta proporcionada (filepath_or_buffer).
* Lee el contenido línea por línea, separando los valores por delimitadores (por defecto, la coma ,).
* Asigna la primera fila como encabezado (header=0 por defecto), a menos que se indique lo contrario.
* Convierte cada columna en una Series de Pandas con un tipo de dato inferido (int, float, string, bool, etc.).
* Devuelve un objeto DataFrame, optimizado para análisis y manipulación en memoria.


#### 📌 Significado detallado:
* 'data/batallas.csv' es la ruta relativa al archivo CSV que contiene los datos de las batallas (por ejemplo: nombre, año, ejércitos, comandantes, resultados, etc.).
* pd.read_csv(...) lee el archivo ubicado en esa ruta y lo convierte en un DataFrame.
* batallas = ... asigna ese DataFrame a la variable batallas, que queda disponible en memoria para posteriores análisis, visualización o modelado.

```python
batallas = pd.read_csv('data/batallas.csv')  # Carga un archivo CSV y lo convierte en un DataFrame de Pandas
```

### 🧠 ¿Qué hace .head() en Pandas?

#### 📌 Descripción técnica:
.head() es un método del objeto DataFrame (o Series) de Pandas que permite visualizar las primeras n filas del conjunto de datos.
* Por defecto, muestra las primeras 5 filas (n=5).
* Es útil para realizar una inspección inicial del contenido, verificar columnas, tipos de datos y estructura general tras cargar o manipular un dataset.

#### 🔧 ¿Qué hace internamente?
* Accede a las primeras n posiciones del índice del DataFrame.
* Retorna una vista parcial de los datos (no modifica el objeto original).
* Optimiza la vista para ser representada de forma legible en consola o notebooks.

```python
batallas.head()  # Muestra las primeras 5 filas del DataFrame para inspección rápida
```

### 🧠 ¿Qué hace .shape en Pandas?

#### 📌 Descripción técnica:
.shape es un atributo (no un método) de los objetos DataFrame y Series en Pandas. Retorna una tupla (filas, columnas) que indica las dimensiones de la estructura de datos.
* 🔧 ¿Qué devuelve exactamente?
    * Un objeto tuple de Python.
    * En un DataFrame, df.shape devuelve:
        * (n_filas, n_columnas)
* En una Series, devuelve solamente el número de elementos (como una tupla de un solo valor).

```python
batallas.shape
```

### 🧠 ¿Qué hace .rename() en Pandas?

#### 📌 Descripción técnica:
.rename() es un método de los objetos DataFrame y Series en Pandas que permite renombrar columnas o índices de manera explícita, sin alterar los datos. Puede devolver un nuevo DataFrame o modificar el original si se usa con inplace=True.
* 🔧 ¿Qué puede renombrar?
    * columns: renombrar los nombres de columnas.
    * index: renombrar etiquetas de filas (índice).
    * inplace: si se pone en True, aplica los cambios sobre el DataFrame original (por defecto es False).


#### 🧪 Caso de uso común en ciencia de datos
* Estándar para limpieza inicial del dataset.
* Útil cuando las columnas tienen nombres inconsistentes, en inglés o mal escritos.
* También para adaptar datasets a reglas específicas de visualización, modelos o APIs.

```python
batallas.rename(columns={'atacante_1':'primer_atacante'},inplace=True)  # Renombra columnas (o índices) de un DataFrame de forma explícita

batallas.head()
```

### 🧠 ¿Qué hace .value_counts() en Pandas?

#### 📌 Descripción técnica:
.value_counts() es un método de objetos Series en Pandas que devuelve una Serie con la frecuencia de cada valor único, es decir, cuántas veces aparece cada valor distinto.
Es una herramienta fundamental para análisis exploratorio de variables categóricas o discretas.

* 🔧 ¿Qué hace internamente?
    * Agrupa los valores únicos de la columna.
    * Cuenta cuántas veces aparece cada valor.
    * Ordena el resultado de mayor a menor (por defecto).
    * Devuelve el resultado como una nueva Series, con el valor como índice y la frecuencia como valor.


* ⚙️ Argumentos útiles:
    * normalize=True: en lugar del conteo absoluto, devuelve la proporción (frecuencia relativa).
    * sort=False: evita el ordenamiento automático.
    * ascending=True: ordena de menor a mayor.
    * dropna=False: incluye los valores NaN en el conteo.

```python
batallas['rey_atacante'].value_counts()  # Cuenta la frecuencia de cada valor único en una columna
```

```python
batallas['ubicacion'].value_counts()
```

### 🧠 ¿Qué hace sns.set_theme()?

#### 📌 Descripción técnica:
sns.set_theme() es una función de configuración global de Seaborn que define el estilo visual y la paleta de colores para todos los gráficos que se generen en adelante en la sesión actual.
* 🔧 ¿Qué modifica?
    * El fondo del gráfico (style: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks').
    * Los colores por defecto (palette: 'deep', 'muted', 'pastel', etc.).
    * Las fuentes, tamaños, bordes y escalas.


### 🧠 ¿Qué hace sns.barplot()?

#### 📌 Descripción técnica:
sns.barplot() genera un gráfico de barras estadístico, que agrega automáticamente los valores usando una función (por defecto np.mean) y representa barras con su intervalo de confianza (ci).
* 🔧 ¿Qué parámetros acepta?
    * x: columna categórica (eje horizontal).
    * y: columna numérica (altura de las barras).
    * data: el DataFrame que contiene los datos.
    * estimator: función agregada (np.mean, np.sum, etc.).
    * ci: intervalo de confianza (95% por defecto, o None para quitarlo).

### 1. .notnull()

🧠 Descripción técnica:
Método de Pandas que devuelve una Serie booleana con True en las posiciones donde el valor no es nulo (NaN).

🔧 ¿Para qué sirve?
Permite filtrar registros válidos en una columna antes de realizar transformaciones que no aceptan NaN, como convertir tipos o aplicar funciones matemáticas.


### 2. .copy()

🧠 Descripción técnica:
Crea una copia independiente del subconjunto del DataFrame. Es crucial para evitar advertencias (SettingWithCopyWarning) cuando se modifican datos después de un filtrado.

⚠️ ¿Por qué es importante?
Si no se usa .copy(), Pandas puede modificar la vista original de decesos, lo que puede generar resultados inesperados.


### 3. .astype(int)

🧠 Descripción técnica:
Convierte explícitamente los datos de la columna al tipo de dato entero (int). Si los datos vienen como float64 (ej: 299.0), esta conversión los transforma en enteros (299).

🔧 ¿Por qué se usa?
Mejora la limpieza del dataset.

Reduce el consumo de memoria.

Evita errores en visualización (por ejemplo, ejes con decimales innecesarios en seaborn).

✅ Requiere que no haya NaN, por eso se hace después de .notnull().


### 4. .unique()

🧠 Descripción técnica:
Devuelve un ndarray de NumPy con todos los valores únicos de una columna (sin duplicados), en el orden en que aparecen.

🔧 ¿Para qué sirve?
Ideal para:

Verificar cuántos años distintos hay.

Saber si los datos ya están limpios.

Usar los valores únicos como etiquetas para visualización, agrupamiento, etc.


### 5. plt.title()

🧠 Descripción técnica:
Asigna un título principal al gráfico actual generado por Matplotlib.

⚙️ Argumentos principales:
label → texto del título (string).

fontsize, fontweight, loc → personalización del estilo del título.


### 7. plt.xlabel()

🧠 Descripción técnica:
Etiqueta el eje horizontal (X) del gráfico con texto personalizado.


### 8. plt.ylabel()

🧠 Descripción técnica:
Etiqueta el eje vertical (Y) del gráfico con una descripción textual.


### 9. plt.tight_layout()

🧠 Descripción técnica:
Ajusta automáticamente el espaciado entre elementos del gráfico (título, ejes, etiquetas, leyendas) para que no se superpongan ni queden cortados, especialmente útil en gráficos con etiquetas largas o rotadas.

🔧 ¿Qué hace internamente?
Recalcula márgenes y padding.

Reorganiza los elementos del gráfico para maximizar legibilidad.

Es indispensable cuando se guarda la figura con plt.savefig() para evitar cortes.
